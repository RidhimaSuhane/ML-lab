```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Step 1: Create Dataset
data = {
    'Age':[22,25,47,52,46,56,23,30,35,40],
    'Salary':[25000,32000,52000,110000,90000,95000,27000,40000,60000,70000],
    'Experience':[1,2,20,25,15,28,1,5,7,10],
    'Purchased':[0,0,1,1,1,1,0,0,1,1]
}

df = pd.DataFrame(data)

print("Dataset:\n")
print(df)

# Step 2: Features and Target
X = df[['Age','Salary','Experience']]
y = df['Purchased']

# Step 3: Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Step 4: Train Random Forest Model
rf = RandomForestClassifier(n_estimators=10, random_state=42)
rf.fit(X_train, y_train)

# Step 5: Prediction
y_pred = rf.predict(X_test)

print("\nTest Data Predictions:")
print(y_pred)

print("\nActual Values:")
print(y_test.values)

# Step 6: Accuracy
accuracy = accuracy_score(y_test, y_pred)
print("\nModel Accuracy:", accuracy)

# Step 7: Feature Importance
print("\nFeature Importance:")
for feature, importance in zip(X.columns, rf.feature_importances_):
    print(feature, ":", round(importance,3))

# Step 8: Predict New Customer
new_data = pd.DataFrame({
    'Age':[45],
    'Salary':[80000],
    'Experience':[12]
})

prediction = rf.predict(new_data)

print("\nNew Customer Data:")
print(new_data)

print("\nPrediction:", prediction)
```
